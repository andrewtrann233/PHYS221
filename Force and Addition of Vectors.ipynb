# Force and Addition of Vectors Lab Analysis

This analysis utilizes the polar coordinate system, so tuples are represented as polar coordinates $(r, \theta)$. Furthermore, forces will be measured in "gram weight" $gmwt$. One gmwt is the force of gravity on one gram of mass.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def components(coordinate: tuple) -> tuple:
    x = coordinate[0] * np.cos(coordinate[1])
    y = coordinate[0] * np.sin(coordinate[1])

    return x, y
```

```python
def resultant(coordinate_1: tuple, coordinate_2: tuple) -> tuple:
    x_1, y_1 = components(coordinate_1)
    x_2, y_2 = components(coordinate_2)

    r = np.sqrt(
        np.square(x_2 - x_1) + np.square(y_2 - y_1)
    )

    adjacent = x_2 - x_1 if x_2 > x_1 else x_1 - x_2
    opposite = y_2 - y_1 if y_2 > y_1 else y_1 - y_2

    theta = np.arctan(
        opposite / adjacent
    )

    return r, theta
```

# Procedure 1

$$
\begin{aligned}
    \vec A &= (100 gmwt, 20\degree) \\
    \vec B &= (200 gmwt, 120\degree) \\
    \vec C &= \vec A + \vec B \\
\end{aligned}
$$

```python
A = (100, np.radians(20))
A_x, A_y = components(A)

B = (200, np.radians(120))
B_x, B_y = components(B)

C = resultant(A, B)
C_x, C_y = components(C)
```

```python
fig, ax = plt.subplots(figsize=(5, 5))

ax.quiver(0, 0, A_x, A_y, angles="xy", scale_units="xy", scale=1, color="r")
ax.quiver(0, 0, B_x, B_y, angles="xy", scale_units="xy", scale=1, color="g")
ax.quiver(0, 0, C_x, C_y, angles="xy", scale_units="xy", scale=1, color="b")

ax.set_title("Procedure 1")
ax.grid()

ax.set_xlabel("x")
ax.axhline(0, linewidth=0.5)
ax.set_xlim(-200, 200)

ax.set_ylabel("y")
ax.axvline(0, linewidth=0.5)
ax.set_ylim(-200, 200)

fig.savefig("Procedure 1.svg")
plt.show()
```
