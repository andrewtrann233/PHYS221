```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
displacement = np.array([2.56, 2.83, 3.14, 3.39, 3.65, 3.99, 4.18, 4.45, 4.80, 5.01, 5.23, 5.57, 5.82, 6.05])
velocity = displacement * 60
total_distance = np.cumsum(displacement)
total_time = np.arange(1, len(displacement) + 1) / 60
```

# Total Distance vs. Time

```python
fig, ax = plt.subplots(figsize=(16, 9))

ax.set_title("Total Distance vs. Time")
ax.grid()

ax.scatter(total_time, total_distance)

coefficients = np.polyfit(total_time, total_distance, 2)
quadratic_model = np.poly1d(coefficients)

ax.plot(total_time, quadratic_model(total_time))

ax.set_xlabel("Time (s)")
ax.set_xlim(0, 0.25)

ax.set_ylabel("Total Distance (cm)")
ax.set_ylim(0, 80)

fig.savefig("Free Fall (Total Distance vs. Time)")
plt.show()
```

# Velocity vs. Time

```python
fig, ax = plt.subplots(figsize=(16, 9))

ax.set_title("Velocity vs. Time")
ax.grid()

ax.scatter(total_time, velocity)

coefficients = np.polyfit(total_time, velocity, 1)
linear_model = np.poly1d(coefficients)

ax.plot(total_time, linear_model(total_time))

ax.set_xlabel("Time (s)")
ax.set_xlim(0, 0.25)

ax.set_ylabel("Velocity (cm/s)")
ax.set_ylim(0, 380)

fig.savefig("Free Fall (Velocity vs. Time)")
plt.show()
```
